## **Linux Basic Directories**
- `/` or "root": The base/top-most directory in Linux **(NOTE: DON'T PLAY WITH THIS IF YOU DON'T KNOW WHAT YOU'RE DOING!!!)**
- `~` or `/home/<username>` or "home": Current user's directory

## **Linux Basic Commands**
- `ls`: List the files and directories in a directory
```bash
ls
ls <target_directory>

# Examples:
# 1. List files and folders in current folder
ls
# 2. List files and folders in /home/username/Downloads folder
ls ~/Downloads
```

- `cd`: Change directory
```bash
cd <destination_directory>

# Examples:
# 1. Change directory to root
cd /
# 2. Change directory to home
cd ~
# 3. Change directory to /home/username/Downloads
cd ~/Downloads
```

- `mkdir`: Create a directory
```bash
mkdir <directory_name>

# Example:
# 1. Create /home/username/pelatihan-genap directory
mkdir ~/pelatihan-genap
```

- `touch`: Create a file
```bash
touch <file_name.extension>

# Examples:
# 1. Create a Python file called "main.py"
touch main.py
# 2. Create a Notebook file called "contoh.ipynb" inside "~/pelatihan-genap" folder
touch ~/pelatihan-genap/contoh.ipynb
```

- `mv`: Move a file (Note: can be used to rename a file)
```bash
# Moving a file
mv <source>/<file_name> <destionation>/<file_name>

# Renaming a file
mv <old_name> <new_name>

# Examples:
# 1. Moving a file
mv ~/Downloads/image-1.png ~/pelatihan-genap/image-1.png
# 2. Renaming a file
mv ~/pelatihan-genap/image-1.png ~/pelatihan-genap/image-1.png
```

- `rm`: Remove a file (NOTE: USE THIS **VERY VERY VERY CAREFULLY**)
```bash
rm <file_name>

# Examples:
# 1. Removing "image-1.png" inside "~/pelatihan-genap" folder
rm ~/pelatihan-genap/image-1.png
# 2. Removing "~/pelatihan-genap" folder and EVERTHING INSIDE
rm -r ~/pelatihan-genap
# 3. Removing an EMPTY DIRECTORY called "~/pelatihan-genap" 
rm -d ~/pelatihan genap
```

### **Splitting The Color Channels**

**Syntax:**
```python
ch_1, ch_2, ..., ch_n = cv2.split(image)
```

````python
# Load the sample image
lake_image = cv2.imread("assets/New_Zealand_Lake.jpg")

# Convert the color space from BGR to RGB
lake_image_corrected = cv2.cvtColor(lake_image, cv2.COLOR_BGR2RGB)

# Split the R, G, and B channel
r, g, b = cv2.split(lake_image_corrected)

# Display each channels
plt.subplot(1, 4, 1)
plt.title("Original")
plt.axis("off")
plt.imshow(lake_image_corrected)

plt.subplot(1, 4, 2)
plt.title("R")
plt.axis("off")
plt.imshow(r, cmap="gray")

plt.subplot(1, 4, 3)
plt.title("G")
plt.axis("off")
plt.imshow(g, cmap="gray")

plt.subplot(1, 4, 4)
plt.title("B")
plt.axis("off")
plt.imshow(b, cmap="gray")

plt.show()
````

### **Merge The Color Channels**

**Syntax:**
```python
merged_image = cv2.merge((ch_1, ch_2, ..., ch_n))
```

````python
merged_lake = cv2.merge((r, g, b))

# Display the image
plt.imshow(merged_lake)
plt.title("Merged Channels")
plt.axis("off")

plt.show()
````

## **Resizing Images**

**Syntax:**
```python
result = cv2.resize(source, dsize, dest, fx, fy, interpolation)
```

- `result`: Resizing result
- `source`: Source image
- `dsize`: Output size
- `fx`: x-scaler
- `fy`: y-scaler
- `interpolation`: Interpolation method
  Examples:
  - `cv2.INTER_NEAREST`: Nearest-neighbor interpolation.
  - `cv2.INTER_LINEAR`: Bilinear interpolation (default).
  - `cv2.INTER_CUBIC`: Bicubic interpolation.
  - `cv2.INTER_LANCZOS4`: Lanczos interpolation over 8×8 neighborhood.

### **Method 1: Using `fx` and `fy`**

````python
# Load the image
futaba_image = cv2.imread("assets/futaba-rio.png")

# Convert the color space from BGR to RGB
futaba_image = cv2.cvtColor(futaba_image, cv2.COLOR_BGR2RGB)

# See the shape before resizing
print(f"Image's shape before resizing: {futaba_image.shape}")

# Resize the image to be 2 times bigger
futaba_image_2x = cv2.resize(futaba_image, dsize=None, fx=2, fy=2)

# See the shape after resizing
print(f"Image's shape after resizing: {futaba_image_2x.shape}")

# Resize the image to be 2 times smaller
futaba_image_half = cv2.resize(futaba_image, dsize=None, fx=0.5, fy=0.5)

# See the shape after resizing
print(f"Image's shape after resizing: {futaba_image_half.shape}")
````

````python
# Resize the image with fx=0.5 and fy=2
futaba_stretched = cv2.resize(futaba_image, dsize=None, fx=0.5, fy=2)

# Display the image
plt.imshow(futaba_stretched)
plt.title("Waifu mu ke-stretch mas")
plt.axis("off")
plt.show()
````

### **Method 2: Specifying The Output Size**

````python
# Load the image
futaba_image = cv2.imread("assets/futaba-rio.png")

# Convert the color space from BGR to RGB
futaba_image = cv2.cvtColor(futaba_image, cv2.COLOR_BGR2RGB)

# See the dimension
print(f"Image's shape before resizing: {futaba_image.shape}")

# Resize the image
futaba_resized = cv2.resize(futaba_image, dsize=(1000, 2000))

# Display the image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(futaba_image)

plt.subplot(1, 2, 2)
plt.title("Resized")
plt.imshow(futaba_resized)

plt.show()
````

## **Flipping Images**

**Syntax:**
```python
result = cv2.flip(source, FLIP_CODE)
```

**`FLIPCODE`:**
- `0`: Flip in x-axis
- `1`: Flip in y-axis
- `-1`: Flip in both axes

````python
# Load a sample image
sample_image = cv2.imread("assets/montir-pesawat.png")

# Convert the color space from BGR to RGB
sample_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)

# Flip the image in x-axis
sample_image_flip_x = cv2.flip(sample_image, 0)

# Flip the image in y-axis
sample_image_flip_y = cv2.flip(sample_image, 1)

# Flip the image in both axes
sample_image_flip_both = cv2.flip(sample_image, -1)

# Display the image
plt.subplot(1, 4, 1)
plt.title("Original")
plt.axis("off")
plt.imshow(sample_image)

plt.subplot(1, 4, 2)
plt.title("Flip X")
plt.axis("off")
plt.imshow(sample_image_flip_x)

plt.subplot(1, 4, 3)
plt.title("Flip Y")
plt.axis("off")
plt.imshow(sample_image_flip_y)

plt.subplot(1, 4, 4)
plt.title("Flip Both")
plt.axis("off")
plt.imshow(sample_image_flip_both)

plt.show()
````

## **Rotating Images**

**Syntax:**
```python
result = cv2.rotate(source, ROTATE_CODE)
```

**`ROTATE_CODE`:**
- `cv2.ROTATE_90_CLOCKWISE`: 90 degrees clockwise
- `cv2.ROTATE_90_COUNTERCLOCKWISE`: 90 degrees counterclockwise
- `cv2.ROTATE_180`: 180 degrees clockwise

````python
# Load a sample image
sample_image = cv2.imread("assets/montir-pesawat.png")

# Convert the color space from BGR to RGB
sample_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)

# Flip the image in x-axis
sample_image_90_cw = cv2.rotate(sample_image, cv2.ROTATE_90_CLOCKWISE)

# Flip the image in y-axis
sample_image_90_ccw = cv2.rotate(sample_image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Flip the image in both axes
sample_image_180 = cv2.rotate(sample_image, cv2.ROTATE_180)

# See the dimensions of each images
print(f"sample_image's shape: {sample_image.shape}")
print(f"sample_image_90_cw's shape: {sample_image_90_cw.shape}")
print(f"sample_image_90_ccw's shape: {sample_image_90_ccw.shape}")
print(f"sample_image_180's shape: {sample_image_180.shape}")

# Display the image
plt.subplot(1, 4, 1)
plt.title("Original")
plt.axis("off")
plt.imshow(sample_image)

plt.subplot(1, 4, 2)
plt.title("Rot 90 CW")
plt.axis("off")
plt.imshow(sample_image_90_cw)

plt.subplot(1, 4, 3)
plt.title("Rot 90 CCW")
plt.axis("off")
plt.imshow(sample_image_90_ccw)

plt.subplot(1, 4, 4)
plt.title("Rot 180")
plt.axis("off")
plt.imshow(sample_image_180)

plt.show()
````
